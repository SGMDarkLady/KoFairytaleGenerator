```python
#유튜브 크롤링을 위한 모듈
from pytube import Playlist
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

import pandas as pd
```

```python
playlist = []
pl = Playlist("https://www.youtube.com/watch?v=B9zF3gIG-c4&list=PLwWDFGSQwdGtlgcRZFRWGdyDPQMzEG9b5")

titles = []
story = []
text = ''

for url in pl:
    key = url.split('=')
    playlist.append(key[1])

for i in range(len(playlist)):
    url = str(playlist[i])
    yt_url = YouTube('https://www.youtube.com/watch?v='+url)
    caption = yt_url.captions.get_by_language_code('a.ko')
    if caption == None:
        i+=1
    else:
        srt = YouTubeTranscriptApi.get_transcript(url, languages=['ko'])
        titles.append(yt_url.title)
        for w in range(len(srt)):
            text += srt[w]['text']+' '
        story.append(text)
        text = ''
        i+=1
```

```python
data = {'제목':titles, '동화':story}
df = pd.DataFrame(data)
df.describe()
```

```python
df.head()
```

```python
df.to_excel('re_ollybolly_story_dataset.xlsx')
df.to_csv('re_ollybolly_story_dataset.csv')
```

# 유튜브 mp3 크롤링

```python
from pytube import Playlist
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

import pandas as pd
```

```python
playlist = []
pl = Playlist("https://www.youtube.com/watch?v=ezWKQI9bTrQ&list=PLZCS0lfVs5MhoIh2lkX0TZqY5NBPKfqrT")

titles = []
story = []
text = ''

for url in pl:
    key = url.split('=')
    playlist.append(key[1])
    
len(playlist)
```

```python
from pytube import YouTube
import glob
import os.path

#유튜브 전용 인스턴스 생성
for i in range(len(playlist)):
    url = str(playlist[i])
    yt_url = YouTube('https://www.youtube.com/watch?v='+url)

    yt_url.streams.filter(only_audio=True).all()

    yt_url.streams.filter(only_audio=True).first().download()
print('success')

# 확장자 변경
files = glob.glob("*.mp4")
for x in files:
    if not os.path.isdir(x):
        filename = os.path.splitext(x)
        try:
            os.rename(x,filename[0] + '.mp3')
        except:
            pass
```

# AI 허브 문학 낭송 음성 원본 데이터

```python
import json

file = open('1.학습레이블.json',encoding='UTF-8')
jsonString = json.load(file)
len(jsonString)
```

```python
origin_data = []
titles = []
for i in range(len(jsonString)):
    data = jsonString[i]
    title = data.get('recite_src').get('literature').get('title')
    text_data = data.get('recite_src').get('text')
    titles.append(title)
    origin_data.append(text_data)
titles
```

```python
data = {'제목':titles, '동화':origin_data}
AI_hub_data = pd.DataFrame(data)
AI_hub_data.describe()
```

```python
AI_hub_data.to_excel('AIhub_dataset.xlsx')
```
