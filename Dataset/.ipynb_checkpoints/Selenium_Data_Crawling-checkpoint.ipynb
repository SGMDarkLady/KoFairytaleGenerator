```python
from selenium import webdriver#동적 구현을 위한 셀리늄 활용
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
import pandas as p

driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('http://18children.president.pa.go.kr/our_space/books.php')
```

```python
import urllib

title = []
story = []
for n in range(1,25):
    html = 'http://18children.president.pa.go.kr/our_space/books.php?srh%5Bcategory%5D=08&srh%5Bpage%5D=' + str(n)
    response = urlopen(html)
    driver.get(html)
    soup = BeautifulSoup(response, "html.parser")
    
    contents = soup.find_all('div', class_='board_list')
    titles = soup.find_all('dt', class_='title')

    for i in range(len(titles)):
        title.append(titles[i].get_text())
    
        num = str(i+1)
        
        detail_search_button = driver.find_element(By.CSS_SELECTOR,"#content > div.real_cont > div.board_list > ul > li:nth-child("+num+") > dl > div > a")
        detail_search_button.click()

        text = driver.find_element(By.CSS_SELECTOR,"#content > div.real_cont > div.board_view > div.box_body > div").text
        story.append(text)

        list_button = driver.find_element(By.CSS_SELECTOR, '#content > div.real_cont > div.buttons > div > a.btn_pack.list')
        list_button.click()
    print(n)
```

```python
print(len(title), len(story))
```

```python
driver.close()
```

```python
data = {'제목':title, '동화':story}
df = pd.DataFrame(data)
df.describe()
```

```python
df.to_excel('bluehouse_2_story_dataset.xlsx')
df.to_csv('[csv]bluehouse_2_story_dataset.csv')
```
